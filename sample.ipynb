```python
df1 = pd.DataFrame(pd.date_range('1/1/2013' , '6/30/2016'), columns=['date'])
```

```python
for j in range(11):
        df2 = pd.DataFrame([df1['date']]*j)
```

```python
df3 = pd.DataFrame(df2.stack())
df = df3.reset_index()
```

```python
df4 = pd.DataFrame(np.random.randint(0, 1000, size = (12770)), columns=['store'])
```

```python
df5 = pd.DataFrame(np.random.randint(0, 2000, size = (12770)), columns=['product'])
```

```python
df6 = pd.DataFrame(np.random.randint(10000, 100000, size = (12770)), columns=['sum'])
```

```python
df7 = pd.DataFrame(np.random.randint(0, 100, size = (12770)), columns=['quantity'])
```

```python
df8 = pd.concat([df4, df5], axis=1)
```

```python
df9 = pd.concat([df8, df6], axis=1)
```

```python
df10 = pd.concat([df9, df7], axis=1)
```

```python
df11 = pd.concat([df10, df], axis=1)
```

```python
df12 = df11.drop("level_0", axis=1)
df13 = df12.drop("level_1", axis=1)
df13.columns = ['store', 'product', 'sum', 'quantity', 'date']
df13
```

```python
df14 = df13.ix[:,['date', 'store', 'product', 'sum', 'quantity']]
df14
```

```python
df14.to_csv('kadai.csv')
```
